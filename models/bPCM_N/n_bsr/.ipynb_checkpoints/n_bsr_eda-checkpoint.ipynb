```python
def load_and_plot(fname='tr_001_001', label='(2p3) 4So',
                  xlim_param_dict = {'left': .01, 'right': 100.0},
                  ylim_param_dict = {'bottom': 1.0, 'top': 1000.0},
                  ylog=True, xlog=True):
    data=np.loadtxt(fname=fname, skiprows=1)
    fig, ax = plt.subplots(1, 1)
    plotter.n_plot_zats(ax, data, label, 'r-',
                        xlim_param_dict = xlim_param_dict,
                        ylim_param_dict = ylim_param_dict,
                        ylog=ylog, xlog=xlog) 
```

```python
load_and_plot()
```
