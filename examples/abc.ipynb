```python
cnx, cursor = nepc.connect(local=False)
```

```python
n_abc = nepc.Model(cursor, "abc")
```

```python
len(n_abc.cs)
```

# Electron-impact electronic excitation of $\text{N}_2$

All electronic excitations/de-excitations of $\text{N}_2$ appear to be from/to the ground electronic state, $\text{N}_2 (\text{X} {}^1\Sigma)$.  

**Question:** *Do we need cross sections for excitations between non-ground electronic states of $\text{N}_2$?*

De-excitation rates are computed using detailed balance.

```python
nepc.model_summary_df(n_abc.filter(specie="N2", process="excitation"))
```

# Electron impact dissociation and dissociative ionization of $\text{N}_2$

The cross sections used for the atomic density evolution equations due to the different dissociation channels from the ground state, $\text{N}_2 (\text{X} {}^1\Sigma)$, seem to all be taken from ref 20 in Angus et al.

It appears that cross sections for electron-impact dissociation (*not* dissociative ionization) from the electronically excited sates of $\text{N}_2 (X, A, B, C)$ states and $\text{N}_2^+ (X, A, B, C)$ states are from ref 22 in Angus et al. and are determined by the weighted-total-cross-section (WTCS) method.

**Question:** *Do we need to compute cross sections for dissociation and dissociative ionization of non-ground electronic states of $\text{N}_2$ and $\text{N}_2^+$ using a more accurate fully quantum method?*

```python
nepc.model_summary_df(n_abc.filter(specie="N2", process="d_ionization_partial"))
```

```python
nepc.model_summary_df(n_abc.filter(specie="N2", process="dissociation_partial"))
```
