```python
import os
import numpy as np
import matplotlib.pyplot as plt
import plotter
import parser
import eda
import pandas as pd
```

```python
os.chdir("/Users/adamson/projects/cs_eda/n/data")
```

```python
elastic=[]
excitation=[]
momentum=[]
effective=[]
ionization=[]
attachment=[]

files = os.listdir()

for filename in files:
    #print('working in {}'.format(filename))
    with open(filename,'r') as fp:
        parsed_file = parser.parse(fp, filename)
        for entry in parsed_file:
            #print('filename: {}, kind: {}'.format(filename, entry['kind']))
            #append_list(entry) #this would be a more elegant approach
            if (entry['kind'] == 'ELASTIC'):
                elastic.append(entry)
            elif (entry['kind']) == 'EXCITATION':
                excitation.append(entry)
            elif (entry['kind']) == 'MOMENTUM':
                momentum.append(entry)
            elif (entry['kind']) == 'EFFECTIVE':
                effective.append(entry)
            elif (entry['kind']) == 'IONIZATION':
                ionization.append(entry)
            elif (entry['kind']) == 'ATTACHMENT':
                attachment.append(entry)
            else:
                print('no place to file kind {}'.format(entry['kind']))
```

```python
print('elastic: {}\nexcitation: {}\nmomentum: {}\neffective: {}\nionization: {}\nattachment: {}'.format(
    len(elastic), len(excitation), len(momentum), len(effective), len(ionization), len(attachment)))
```

FIG 1 from <cite data-cite="6578527/E3SGE2N7"></cite> (elastic electron scattering from nitrogen atoms in the ($2s^22p^3)\,{}^4S^o$ ground state):

<img src="lit/e-n-elastic-2s2_2p3-4So.png" width="300">

The 0th elastic scattering cross section is for the $(2s^2 2p^3)$ $^4S^o$ state (FIG 1 in the paper). The 1st elastic cross section is the momentum-transfer cross section for this electronic state and is noted with the comment "momentum transfer."

```python
fig, ax = plt.subplots(1, 1)
plotter.lxcat_plot_zats(ax, elastic[0:2], ['r-', 'b--'],
                        plot_param_dict = {'linewidth':.8}, 
                        xlim_param_dict = {'left': 0.01, 'right': 120.0}, 
                        ylog=True, xlog=True) 
```
